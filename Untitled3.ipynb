```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Professionally Presentable

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['font.size'] = 10 # Corrected from font_size to font.size
print("Analysis of the datasets for Future Interns Internship")

# Loading & Cleaning data

df = pd.read_csv("/content/small-business-sales-performance-dataset/sales_data.csv") # Corrected path to include filename

df['Date'] = pd.to_datetime(df['Date'],errors='coerce')
df = df.dropna(subset = ['Date'])
df['Month'] = df['Date'].dt.to_period('M')
df['Year'] = df['Date'].dt.year
df['Month_name'] = df['Date'].dt.strftime('%B %Y')

# Key Metrices calculations

df['Revenue'] = df['Quantity'] * df['Price']
df['Profit'] = df['Revenue'] - df['Quantity'] * df['Cost']
df['Profit_Margin'] = (df['Profit'] / df['Revenue']) * 100
print( f"Loading successful {len(df):,} analyzed.\n")

#  Key Performance Indications

total_revenue = df['Revenue'].sum()
total_profit = df['Profit'].sum()
overall_margin = (total_profit / total_revenue) * 100
avg_order_value = df['Revenue'].mean()
avg_profit_per_order = df['Profit'].mean()
total_orders = len(df)
print("="*60)
print("Key Bussiness Metrics Report")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Overall Profit Margin: {overall_margin:.2f}%")
print(f"Average Order Value: ${avg_order_value:,.2f}")
print(f"Total orders : {total_orders:,}")

# Visualizations

monthly = df.groupby('Month').agg({'Revenue':'sum', 'Profit':'sum', 'Profit_Margin':'mean'}).reset_index()
monthly['Month_name'] = monthly['Month'].dt.strftime('%B %Y')
plt.figure(figsize=(12,6))
plt.plot(monthly['Month_name'], monthly['Revenue'], label='Revenue', marker='o', linewidth =3 , color = '#1f77b4')
plt.plot(monthly['Month_name'] , monthly['Profit'], label='Profit', marker='o', linewidth =3 , color = '#ff7f0e')
plt.title('Monthly Revenue and Profit' , fontsize = 14 , fontweight = 'bold')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.legend()
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()

top_products = df.groupby('Product')['Revenue'].sum().nlargest(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values , y=top_products.index , palette = 'viridis')
plt.title('Top 10 Selling Products by Revenue', fontsize=14,fontweight='bold')
plt.xlabel('Revenue ($)')
plt.tight_layout()
plt.show()

category_per = df.groupby('Category').agg(Revenue=('Revenue', 'sum'), Profit=('Profit', 'sum'), Profit_Margin=('Profit_Margin', 'mean')).round(2)
print("Category Performance Report")
print(category_per)
plt.figure(figsize=(12,6))
sns.barplot(data=category_per.reset_index(),x='Category' , y='Profit_Margin', palette = 'coolwarm')
plt.title('Average Profit Margin by Category', fontsize=14, fontweight='bold')
plt.ylabel('Profit Margin (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Prediction for the future according to the data

print("\n" + "="*60)
print("Key Insights,Predictions and Recommendation")
print("="*60)

top_category = category_per['Revenue'].idxmax()
top_category_rev_share = (category_per.loc[top_category,'Revenue']/total_revenue)*100

print(f"Overall Performace : The small business generated ${total_revenue:,.0f} in revenue with a healthy {overall_margin:.1f}% profit margin.This indicates strong operational efficiency for a small business")
print(f"Category Leadership : {top_category} is the star performer, contributing {top_category_rev_share:.1f}% profit margin. Focus marketing and inventory")
print(f"Trend analysis: Review the 'Monthly Revenue $ Profit Trends' chart. Look for seasonal or growth patterns - if profit is lagging revenue,inventigate cost inflation.")
print(f"Product concentration Risk : Top products likely drive>70% of revenue(check ' Top 10 products chart'). Diverify the portfolio to avoid or reduce risk")
print(f"Margin optimization opportunity : Categories with Proit Margin < Industry average(typically 25-40% for small retail need immediate attention - consider price increase, suplier negociation or discontunuation.")
```
