```python
# Task 3 of Future Interns
# Bank Marketing Campaign Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Importing and Cleaning the datasets

# Fix: Read CSV with semicolon separator
df = pd.read_csv("/content/bank-marketing-campaigns-dataset/bank-additional-full.csv", sep=';')
print(f"Dataset loading complete {df.shape[0]:,} rows * {df.shape[1]}columns")
print("Columns:", df.columns.tolist())
df = df.copy()

# The following columns ('Opening_date', 'DOB', 'Monthly_Income', 'Existing_EMI')
# are not present in the 'bank-additional-full.csv' dataset.
# The original code attempted to create 'Month', 'Year', and 'Age' from 'Opening_date' and 'DOB'.
# The dataset already has an 'age' column and a 'month' column.

# Commenting out lines that rely on non-existent columns for now.
# df['Opening_date'] = pd.to_datetime(df['Opening_date'],errors ='coerce')
# df['DOB'] = pd.to_datetime(df['DOB'],errors ='coerce')
# df['Month'] = df['Opening_date'].dt.month_name()
# df['Year'] = df['Opening_date'].dt.to_period('M').astype(str)
# df['Age'] = (df['Opening_date'] - df['DOB']).dt.days / 365.25 # Calculate age in years

# Use existing 'age' column
df['Age'] = df['age']

# 'Monthly_Income' and 'Existing_EMI' are not in this dataset.
# df['Monthly_Income'] = df['Monthly_Income'].fillna(df['Monthly_Income'].median())
# df['Existing_EMI'] = df['Existing_EMI'].fillna(0)

# Create 'Contacted' column based on 'contact'
# Assuming any known contact type ('cellular', 'telephone') means 'Y', otherwise 'N'.
df['Contacted'] = df['contact'].apply(lambda x: 'Y' if x in ['cellular', 'telephone'] else 'N')

# Create 'Approved' column from 'y' (target variable: 'yes'/'no')
df['Approved'] = (df['y'] == 'yes').astype(int)

df['Contacting'] = df['Contacted'].fillna('N') # This line can stay, but may not have practical effect if Contacted is always 'Y' or 'N'

print("Structured $ Cleaned Data")

total_customers = len(df)
contacted_customers = df[df['Contacted']=='Y'].shape[0]
Approved_customers = df[df['Approved']==1].shape[0]

funnel = pd.DataFrame({'Stage':['Total Customers', 'Contacted Customers', 'Approved Customers'],'Count':[total_customers, contacted_customers, Approved_customers],'Conversion': [100,round(contacted_customers/total_customers*100,2),round(Approved_customers/total_customers*100,2)]})
print("Overall Funnel")
print(funnel.to_string(index=False))
funnel['Drop_down'] = [0] + [round(100 - funnel['Conversion'].iloc[i], 2) for i in range(1,len(funnel))]
print(funnel[['Stage', 'Count', 'Conversion', 'Drop_down']])

# 'Source', 'Source_Category', and 'ID' are not available in this dataset.
# Using 'job' as a proxy for 'Source' and 'poutcome' as a proxy for 'Source_Category'.
# Creating a dummy 'ID' column from the dataframe index.
df['ID'] = df.index # Create a dummy ID column

channel_per = df.groupby('job').agg(
    Leads = ('ID', 'count'),
    contacted = ('Contacted',lambda x: (x=='Y').sum()),
    Approved = ('Approved', 'sum')
).reset_index()
channel_per = channel_per.rename(columns={'job': 'Source'}) # Rename for consistency with original code's variable names
channel_per['Traffic_customers'] = 100
channel_per['Customers__contacted'] = round(channel_per['contacted']/channel_per['Leads']*100,2)
channel_per['Customers_Approved'] = round(channel_per['Approved']/channel_per['Leads']*100,2)
channel_per['Overall_conversion'] = round(channel_per['Approved']/channel_per['Leads']*100,2)
print("Performance by Channel (using 'job' as proxy)")
print(channel_per.sort_values(by='Overall_conversion', ascending=False)[['Source', 'Leads', 'contacted', 'Approved', 'Traffic_customers', 'Customers__contacted', 'Customers_Approved', 'Overall_conversion']])

Category_per = df.groupby('poutcome').agg(
    Leads = ('ID', 'count'),
    Approved = ('Approved', 'sum')
).reset_index()
Category_per = Category_per.rename(columns={'poutcome': 'Source_Category'}) # Rename for consistency
Category_per['Conversion'] = round(Category_per['Approved']/Category_per['Leads']*100,2)
print("Performance by Campaign (using 'poutcome' as proxy)")
print(Category_per.sort_values('Conversion',ascending=False))

# For 'Month_Year', the dataset has 'month' (e.g., 'may', 'jun') but no year information.
# Using 'month' directly and assuming it represents monthly performance without year context.
df['Month_Year'] = df['month'] # Use the existing 'month' column
# To sort correctly, convert to categorical with ordered months.
month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
df['Month_Year'] = pd.Categorical(df['Month_Year'], categories=month_order, ordered=True)

time_per = df.groupby('Month_Year').agg(
    Leads = ('ID', 'count'),
    Approved = ('Approved', 'sum')
).reset_index()
time_per['Conversion'] = round(time_per['Approved']/time_per['Leads']*100,2)
print("Performance by Time (using 'month' as proxy for time periods)")
print(time_per.sort_values('Month_Year'))

# Visualizations of data $ Recommendation through it

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
plt.barh(funnel['Stage'],funnel['Count'],color = ['#1f77b4','#ff7f0e','#2ca02c'])
for i ,v in enumerate(funnel['Count']):
  plt.text( v + 500 ,i, f"{v:} ({funnel['Conversion'].iloc[i]}%)",va='center')
plt.title('Marketing Funnel-OverallConversion flow')
plt.xlabel('Number of records')
plt.gca().invert_yaxis()
plt.show()

plt.figure(figsize=(10,6))
sns.heatmap(channel_per.set_index('Source')[['Customers__contacted', 'Customers_Approved']].T, annot = True , cmap = 'RdYlGn',fmt ='.1e')
plt.title('Stage Conversion by channel (using job as Source)')
plt.show()

plt.figure(figsize=(10,6))
sns.lineplot(data=time_per , x='Month_Year', y ='Conversion', marker = 'o')
plt.title('Performance by Time (using month)')
plt.xticks(rotation = 45)
plt.ylabel('Conversion')
plt.show()

print("=")
```
